```python
df.head()
```

```python
def convert_quarter_to_date(qtr_str):
    quarter, year = qtr_str.split()
    month = {'Q1': 1, 'Q2': 4, 'Q3': 7, 'Q4': 10}[quarter]
    return pd.Timestamp(year=int(year), month=month, day=1)

# Apply the function to create a new datetime column
df['Date'] = df['Date'].apply(convert_quarter_to_date)
```
