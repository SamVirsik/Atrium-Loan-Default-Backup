```python
grouped = df.groupby("Loan Number")["Modification Indicator"]

# Count loans where it's "Y" at least once
at_least_once_Y = grouped.apply(lambda x: (x == "Y").any()).sum()

# Count loans where it's always "Y"
always_Y = grouped.apply(lambda x: (x == "Y").all()).sum()

print("Loans where 'Modification Indicator' is Y at least once:", at_least_once_Y)
print("Loans where 'Modification Indicator' is always Y:", always_Y)
```

```python
modified_loans = df[df['Modification Indicator'] == 'Y']
modified_loans
```

```python
# Group by Loan Number and check modification counts
mod_counts = df.groupby('Loan Number')['Modification Indicator'].value_counts().unstack(fill_value=0)

# Keep only loans with both 'Yes' and 'No' (or NaN) in their history
mixed_mod_loans = mod_counts[(mod_counts.get('Y', 0) > 0) & (mod_counts.get('N', 0) > 0)].index

# Filter full dataset for those loans
mixed_mod_df = df[df['Loan Number'].isin(mixed_mod_loans)]
```
